```python
import pandas as pd
```

```python
df1 = pd.DataFrame({'key1' : [1, 2, 3, 4],
                    'key2' : [5,6,7,8],
                    'key3' : [9,10, 11, 12]})
                   
```

```python
df1
```

```python
df2 = pd.DataFrame({'key1' : [1, 1, 1, 1],
                    'key4' : [4, 4, 4 ,4],
                    'key5' : [5, 5, 5, 5]})
                 
```

```python
df2
```

                                                                                MERGE()

MERGE WORKS On COLOUMN VALUE

```python
pd.merge(df1, df2)
```

```python
pd.merge(df1, df2, how = 'outer')
```

```python
pd.merge(df1, df2, how = 'left')
```

```python
pd.merge(df1, df2, how = 'cross')
```

                                                                            join

JOIN WORKS ON INDEXES 

```python
df1 = pd.DataFrame({'key1' : [1, 2, 3, 4],
                    'key2' : [5,6,7,8],
                    'key3' : [9,10, 11, 12]
                   }, index = ['a', 'b', 'c', 'd']
                  
                  )


df2 = pd.DataFrame({'key6' : [1, 1, 1, 1],
                    'key4' : [4, 4, 4 ,4],
                    'key5' : [5, 5, 5, 5]
                   }, index = ['e', 'f', 'g', 'h']
                  )
                 
```

```python
df1
```

```python
#by default left join

df1.join(df2)
```

```python
#outer join

df1.join(df2, how = 'outer')
```

```python
df1.join(df2, how = 'cross', on = 'key1')
```
