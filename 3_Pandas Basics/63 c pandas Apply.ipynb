```python
import pandas as pd 
```

```python
s = pd.read_html("https://github.com/datasciencedojo/datasets/blob/master/titanic.csv")
```

```python
df = s[0]
```

```python
df
```

CONVERT THE 'Fare' from Euro to inr

```python
#inside function

df['Fare_INR'] = df['Fare'].apply( lambda x : x*87)
```

```python
df
```

```python
#outside function 

def taka(n):
    return n*113

df['Fare_Taka'] = df['Fare'].apply(taka)
```

```python
df
```

Store the length of name in a new coloumn

```python
def length(s):
    return len(s)

df['length'] = df['Name'].apply(length)
```

```python
df
```

If Fare < 10 : cheap , elif Fare <20 : mid else : high

```python
def value(n):
    if n<10:
        return 'cheap'
    
    elif n<20:
        return 'mid'
    
    else: 
        return 'high'
    
    
```

```python
df['Fare Value'] = df['Fare'].apply(value)
```
