### Data Cleaning

```python
#columns which has null values

df.isnull().sum()
```

```python
df.size
```

```python
df = df.dropna()
```

```python
df.isnull().sum()
```

```python
df.size
```

```python
df.iloc[[122]]
```

```python
#remove this 122

df= df.drop(124).reset_index(drop=True)
df.size
```

```python
df.columns
```

```python
#fix the spaces in Column

df.columns = df.columns.str.strip()
df.columns
```

```python
#changing required columns in int data type

df[['month', 'day', 'year', 'Temperature','RH', 'Ws']] = df[['month', 'day', 'year', 'Temperature','RH', 'Ws']].astype(int)
```

```python
df[['Rain', 'FFMC','DMC', 'DC', 'ISI', 'BUI', 'FWI']] = df[['Rain', 'FFMC','DMC', 'DC', 'ISI', 'BUI', 'FWI']].astype(float)
```

```python
df.head(2)
```

```python
df.info()
```

```python
df.loc[:122,'Region']=1
df.loc[122:,'Region']=2
df.tail()
```

```python
df[['Region']] = df[['Region']].astype(int)
df.head(1)
```

```python
### Data Cleaning in Target Variable 

df['Classes'].value_counts()
```

```python
## handle misplace values

df['Classes'] = df['Classes'].str.strip()
df['Classes'].value_counts()
```
