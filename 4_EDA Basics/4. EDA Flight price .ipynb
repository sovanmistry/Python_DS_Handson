### Working with Date

```python
df['Date_of_Journey'].str.split('/')
```

```python
#creating Date 

df['Date'] = df['Date_of_Journey'].str.split('/').str[0]
df['month'] = df['Date_of_Journey'].str.split('/').str[1]
df['year'] = df['Date_of_Journey'].str.split('/').str[2]

df.head(2)
```

```python
#converting new d m y into y

df['Date'] = df['Date'].astype(int)
df['month'] = df['month'].astype(int)
df['year'] = df['year'].astype(int)

df.info()
```

```python
# remove 'Date_of_Journey' coloumn

df.drop(['Date_of_Journey'], axis= 1, inplace = True)
```

```python
df.head(2)
```

```python
#working of Arrival Time


df['Arrival_Time'].str.split(' ').str[0]
```

```python
df['Arrival_time'] = df['Arrival_Time'].str.split(' ').str[0]
```

```python
df.head(2)
```

```python
#Create a hour column for arrival hour and min

df['Arrival_hour'] = df['Arrival_time'].str.split(':').str[0]
df['Arrival_minute'] = df['Arrival_time'].str.split(':').str[0]
df.head(2)
```

```python
#delete 2 column of arrival time

df.drop(['Arrival_time'], axis = 1, inplace = True)
df.drop(['Arrival_Time'], axis = 1, inplace = True)
```

```python
df.head(2)
```

```python
df.info()
```

```python
#convertining Arrival hour and minute in int

df['Arrival_hour'] = df['Arrival_hour'].astype(int)
df['Arrival_minute'] = df['Arrival_hour'].astype(int)
df.info()
```

```python
#Working on Departure time


df['Dep_hour'] = df['Dep_Time'].str.split(':').str[0]
df['Dep_minute'] = df['Dep_Time'].str.split(':').str[1]
df.head()
```

```python
df['Dep_hour'] = df['Dep_hour'].astype(int)
df['Dep_minute'] = df['Dep_minute'].astype(int)
df.drop(['Dep_Time'], axis =1, inplace = True)
df.head(2)
```

```python
df.drop(['Route'], axis = 1, inplace = True)
df.head(2)
```

```python
#Create a duration

df['Duration_hour'] = df['Duration'].str.split('h').str[0]
df['Duration_minute'] = df['Duration'].str.split('h').str[1]
df['Duration_minute'] = df['Duration_minute'].str.split('m').str[0]
df.head(2)
```
