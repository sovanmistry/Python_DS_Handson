```python
from sklearn.datasets import make_classification
```

```python
#x = Independent Feautre - input
# y = Dependent Feature - output (0, 1)
```

```python
x,y = make_classification(n_samples=1000, n_features=2, n_redundant=0, n_clusters_per_class=1, weights=[0.9], random_state=1)
```

```python
x
```

```python
y
```

```python
#converting into data frame

import pandas as pd
df1 = pd.DataFrame(x, columns=['f1', 'f2']) # input DataFrame
df2 = pd.DataFrame(y, columns=["target"])    # target DataFrame
                
final_df = pd.concat([df1, df2],axis=1) #azis =1 column wise arrangement
final_df.head()
```

```python
#check the value count of target is it immbalanced or not

final_df['target'].value_counts()
```

```python
#plot for visual Representation

import matplotlib.pyplot as plt
plt.scatter(final_df['f1'], final_df['f2'], c= final_df['target'])
```

```python
from imblearn.over_sampling import SMOTE 
```

```python
oversample = SMOTE()
```

```python
new_x, new_y = oversample.fit_resample(final_df[['f1', 'f2']], final_df['target'])
```

```python
new_x
```

```python
new_y
```

```python
new_x.shape
```

```python
new_y.shape
```

```python
df1_new = pd.DataFrame(new_x, columns=['f1', 'f2']) # input DataFrame
df2_new = pd.DataFrame(new_y, columns=["target"])    # target DataFrame
                
oversampled_df = pd.concat([df1_new, df2_new],axis=1) #azis =1 column wise arrangement
oversampled_df.head()
```

```python
#representation of oversampled Df 

plt.scatter(oversampled_df['f1'], oversampled_df['f2'], c= oversampled_df['target'])
```
