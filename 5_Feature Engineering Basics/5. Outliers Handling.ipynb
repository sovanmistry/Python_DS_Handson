                                                    Outliers Handling

```python
import numpy as np

lst_marks = [23,43,21,45,33,44,38, 67, 65,54,78, 45, 34,73,58, 89, 1000, 1040]
```

```python
#manually fetc quntiers

Q1 = np.percentile(lst_marks, [25]) #25 percentile 
Q1
```

```python
#shortcut approaach to get 5 no summary

minimum, Q1, Q2, Q3, maximum = np.quantile(lst_marks, [0, 0.25, 0.50, 0.75, 1.0])
minimum, Q1, Q2, Q3, maximum

```

```python
# calculate IQR

IQR = Q3-Q1
IQR
```

```python
#lower Fence & Higher Fence

lower_fence = Q1 - 1.5*IQR
higher_fence = Q3 + 1.5*IQR
lower_fence, higher_fence
```

```python
outliers = []
without_outlier = []
for i in lst_marks:
    if i>=lower_fence and i<higher_fence:
        without_outlier.append(i)
    else:
        outliers.append(i)
```
