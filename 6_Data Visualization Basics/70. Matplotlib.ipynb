```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x = np.linspace(0,10, 200)
```

```python
y = np.sin(x)
```

```python
plt.plot(x,y)
```

```python
plt.ylabel("this is my y data")
plt.xlabel("this is my x data")
plt.title("this is my title")
plt.plot(x, y)
plt.show()
```

```python
x = np.random.rand(50)
y = np.random.rand(50)
```

```python
size = 300*  np.random.rand(50)
color = np.random.rand(50)


plt.scatter(x,y, c= color, s=size , alpha=0.4)
plt.xlabel("this is x axis")
plt.ylabel("this is y axis")
plt.title("title")
```

```python
x = ['a', 'b', 'c', 'd', 'e']
y = np.random.rand(5)
```

```python
plt.bar(x,y)
plt.xlabel("this is x axis")
plt.ylabel("this is y axis")
plt.title("title")
```

```python
#horizontal Bar plot

plt.barh(x,y)
```

```python
x = np.linspace(0,10, 200)
y = np.sin(x)
```

```python
plt.figure(figsize=(10,3))
plt.grid()
plt.plot(x,y, '--b')
```

```python
x = np.random.rand(50)
y = np.random.rand(50)
z = np.random.rand(50)

fig = plt.figure()
ax = fig.add_subplot(projection = '3d')
ax.scatter(x, y, z)
```
