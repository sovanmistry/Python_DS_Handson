                                                TIPS Dataset

```python
tips = sns.load_dataset('tips')
```

```python
tips
```

```python
#check wherter highest bill gives the highest tip ?

sns.scatterplot( x = tips['total_bill'], y = tips['tip'])
```

```python

sns.countplot(x = tips['smoker'])
```

```python
sns.scatterplot( x = tips['total_bill'], y = tips['tip'], hue = tips['smoker'])
```

```python
sns.relplot( x = tips['total_bill'], y = tips['tip'], hue = tips['smoker'])
```

```python
sns.relplot( x = tips['total_bill'], y = tips['tip'], hue = tips['size'])
```

```python
sns.relplot( x = tips['total_bill'], y = tips['tip'], style= tips['size'])
```

```python
sns.relplot( x = tips['total_bill'], y = tips['tip'], style= tips['size'], hue = tips['time'])
```

```python
#which day has more rusk


sns.catplot(y = tips['total_bill'], x = tips['day'])
```
