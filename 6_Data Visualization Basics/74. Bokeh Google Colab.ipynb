```python
from bokeh.plotting import figure, output_file, show

#bokeh has default dataset
from bokeh.sampledata.iris import flowers

#output file
output_file('test.html')

p = figure(title = "Test Flower from Bokeh")
p.xaxis.axis_label = "petal_length"
p.yaxis.axis_label = "petal_width"
p.circle(flowers['petal_length'], flowers['petal_width'])
#flowers.head()
show(p)
```

```python
#bokeh line plotting
from bokeh.plotting import figure, output_file, show

#output file
output_file('test.html')

p = figure(title = "Test Line Bokeh")
p.xaxis.axis_label = "X axis"
p.yaxis.axis_label = "Y axis"

x = [1,2,3,4]
y = [12,2,3,11]
#p.line(x,y)
p.scatter(x,y, color = 'red', legend_label = 'Red Circle', size = 20)
#flowers.head()
show(p)
```
